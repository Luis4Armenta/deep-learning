## 1. Cargar el conjunto de datos

```python
from tensorflow.keras.datasets import reuters
```

```python
(train_data, train_labels), (test_data, test_labels) = reuters.load_data(num_words=10000)
```

El argumento num_words indica que únicamente tendremos en cuenta las 10,000 palabras más frecuentes del conjunto de datos. Por lo tanto el resto de palabras que sean frecuentes, serán descartadas y no formaran parte de las reseñas con las que entrenaremos y probaremos el algoritmo

Tanto train_data como test_data se corresponden a una lista de noticias, Cada una de las notificias se corresponde con una lista de números enteros que representan el indice de una palabra. Para conocer qué índice tiene cada palabra, podemos llamar el método get_word_index()

## Preprocesamiento del conjunto de datos

A pesar de que ya tenemos codificadas las reseñas para que esten que estén representadas por valores númericos en lugar de por cadenas de texto, existen algunos factores que no permiten que sean conjuntos de datos apropiados para proporcionarle a una red neuronal artificial.

```python
import numpy as np

def vectorize(seqs, dim=10000):
    results = np.zeros((len(seqs), dim))
    for i, seq, in enumerate(seqs):
        results[i, seq] = 1.
    return results
```

Vectorizamos tanto el conjunto de datos de entrenamiento, como el conjunto de datos de pruebas.

```python
# Vectorizamos el conjunto de datos de entrenamiento y el de pruebas
X_train = vectorize(train_data)
X_test = vectorize(test_data)
```

```python
# Cada una de las reseñas se corresponde con un vector formado por 1s y 0s
print('Valores originales:\t', train_data[0][:15])
print('Valores vectorizados:\t', X_train[0][:15])
```

```python
print(str(X_train[0][6]))
```

```python
# Transformamos las etiquetas de salida en vectores
from tensorflow.keras.utils import to_categorical

Y_train = to_categorical(train_labels)
Y_test = to_categorical(test_labels)
```

```python
train_labels[0]
```

```python
Y_train[0]
```

## División del conjunto de datos

```python
from sklearn.model_selection import train_test_split

X_test, X_val, Y_test, Y_val = train_test_split(X_test, Y_test, test_size=0.5)
```

```python
print('Logitud subconjunto de entrenamiento: ', len(X_train))
print('Logitud subconjunto de validación: ', len(X_val))
print('Logitud subconjunto de pruebas: ', len(X_test))
```

## Construcción de la red neuronal artificial

```python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(46, activation='softmax'))
```

```python
from tensorflow.keras import optimizers
opt_func = optimizers.RMSprop()
```

```python
model.compile(
    optimizer=opt_func,
    loss='categorical_crossentropy',
    metrics=['accuracy', 'Precision']
)
```

```python
len(X_train) / 512
```

```python
history = model.fit(
    X_train,
    Y_train,
    epochs=10,
    batch_size=1024,
    validation_data=(X_val, Y_val)
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)[['loss', 'val_loss']].plot(figsize=(10, 6))
plt.grid(True)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history)[['accuracy', 'val_accuracy']].plot(figsize=(10, 6))
plt.grid(True)
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```

## Predicción con el subconjunto de datos de pruebas

```python
error, accuracy, precision = model.evaluate(X_test, Y_test)
```

```python
print('Error: ', error)
print('Accuracy: ', accuracy)
print('Precision: ', precision)
```
